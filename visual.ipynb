```python


import matplotlib.pyplot as plt
def plot(values, export = True, cl = 'red'):
    ax = plt.subplot(111)
    ax.set_title('The convergence of HMAGS and GA')
    ax.set_xlabel('Generation')
    ax.set_ylabel('Tour-length')
    ax.plot(values, color = cl)
    if export:
        plt.savefig("conv_1.pdf",bbox_inches='tight')    
        plt.show()

f_1 = open('Data/conv_file_1.txt', 'r')
f_2 = open('Data/conv_file_2.txt', 'r')
string = f_1.read()
X_1 = list(map(float, string.split()))
string = f_2.read()
X_2 = list(map(float, string.split()))
while(len(X_1) > 650):
    X_1.pop()
while(len(X_2) > 650):
    X_2.pop()
plot(X_1, False, 'blue')
plot(X_2, True, 'red')
```
